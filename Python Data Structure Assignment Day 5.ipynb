### Create a string, sent = ‚ÄòCoronavirus Caused Lockdowns Around The World.‚Äù

```python
# Creating a string variable with a sentence
sent = 'Coronavirus Caused Lockdowns Around The World.'
print(sent)
```

Use split function to convert the string into a list of words and save it in variable words and print the same

```python
# split() method splits a string into a list of words
# By default, it splits on whitespace (spaces)
words = sent.split()
print(words)  # Output: ['Coronavirus', 'Caused', 'Lockdowns', 'Around', 'The', 'World.']
```

Convert each word in the list to lower case and store it in variable words_lower. 
Print words_lower

```python
# Using list comprehension to convert each word to lowercase
# lower() method converts string to lowercase
words_lower = [word.lower() for word in words]
print(words_lower)  # All words are now in lowercase
```

Check whether ‚Äòcountry‚Äô is in the list

```python
# 'in' keyword checks if an element exists in the list
# Returns True if found, False otherwise
print('country' in words_lower)  # Output: False (country is not in the list)
```

Remove the element ‚Äòthe‚Äô from the list and print the list.

```python
# remove() removes the first occurrence of the specified element
# Note: 'the' exists in words_lower (converted from 'The')
words_lower.remove('the')
print(words_lower)  # 'the' is now removed from the list
```

Select the first 4 words from the list words_lower using slicing and store them in a new variable x4

```python
# Slicing: list[start:end] - end is exclusive
# [:4] means from index 0 to 3 (first 4 elements)
x4 = words_lower[:4]
```

```python
# print x4

# Printing the first 4 words from the list
print(x4)  # Output: first 4 words
```

Convert the list of elements to single string using join function and print it

```python
# join() method combines list elements into a single string
# ' '.join(list) - uses space as separator between elements
joined_string = ' '.join(x4)
print(joined_string)  # Output: words joined by spaces
```

## Sets

### Create stud_grades = ['A','A','B','C','C','F']

```python
# Creating a list with duplicate grade values
# Lists allow duplicate elements
stud_grades = ['A', 'A', 'B', 'C', 'C', 'F']
print(stud_grades)
```

Print the len of stud_grades

```python
# len() counts total elements including duplicates
print(len(stud_grades))  # Output: 6
```

Create a new variable, stud_grades_set = set(stud_grades)

```python
# set() converts a list to a set, removing all duplicates
# Sets only store unique elements
stud_grades_set = set(stud_grades)
```

Print stud_grades_set. 

```python
# Printing the set - notice duplicates are removed
# Set order may vary as sets are unordered
print(stud_grades_set)  # Output: {'A', 'B', 'C', 'F'}
```

print the type of stud_grades and stud_grades_set and print their corresponding elements. Try to understand the difference between them.

```python
# Comparing list and set
print("Type of stud_grades:", type(stud_grades))      # <class 'list'>
print("Type of stud_grades_set:", type(stud_grades_set))  # <class 'set'>

print("\nList elements (with duplicates):", stud_grades)
print("Set elements (unique only):", stud_grades_set)
# Notice: List has 6 elements, Set has only 4 (duplicates removed)
```

Add a new element ‚ÄòG‚Äô to stud_grades_set

```python
# add() method adds a new element to the set
# Sets use add() instead of append()
stud_grades_set.add('G')
print(stud_grades_set)  # 'G' is now added to the set
```

Add element 'F' to stud_grades_set. and print it.

```python
# Trying to add 'F' which already exists
# Sets ignore duplicates - 'F' won't be added again
stud_grades_set.add('F')
print(stud_grades_set)  # No change - 'F' was already present
```

!!Did you notice? set doesn't add an element if it's already present in it, unlike lists.

Remove ‚ÄòF‚Äô from stud_grades_set

```python
# remove() removes a specific element from the set
# Raises KeyError if element doesn't exist
stud_grades_set.remove('F')
print(stud_grades_set)  # 'F' is now removed
```

Print the elements and the length of stud_grades_set

```python
# Printing elements and length of the set
print("Elements:", stud_grades_set)
print("Length:", len(stud_grades_set))
```

### Create colors = [‚Äòred‚Äô,‚Äôblue‚Äô,‚Äôorange‚Äô], and fruits = [‚Äòorange‚Äô,‚Äôgrapes‚Äô,‚Äôapples‚Äô] 

```python
# Creating two lists - one with colors, one with fruits
# Notice 'orange' appears in both (it's both a color and a fruit!)
colors = ['red', 'blue', 'orange']
fruits = ['orange', 'grapes', 'apples']
```

Print color and fruits

```python
# Printing both lists
print("Colors:", colors)
print("Fruits:", fruits)
```

Create colors_set, and fruits_set. (using set() ) and print them

```python
# Converting lists to sets for set operations
colors_set = set(colors)
fruits_set = set(fruits)

print("Colors Set:", colors_set)
print("Fruits Set:", fruits_set)
```

Find the union of both the sets.

```python
# union() - combines all elements from both sets (no duplicates)
# Can also use | operator: colors_set | fruits_set
union_result = colors_set.union(fruits_set)
print("Union:", union_result)  # All unique elements from both sets
```

Find the intersection of both the sets 

```python
# intersection() - finds common elements in both sets
# Can also use & operator: colors_set & fruits_set
intersection_result = colors_set.intersection(fruits_set)
print("Intersection:", intersection_result)  # Output: {'orange'}
```

Find the elements which are Fruits but not colors (using set.difference() )

```python
# difference() - elements in first set but NOT in second set
# fruits_set.difference(colors_set) = fruits that are not colors
difference_result = fruits_set.difference(colors_set)
print("Fruits but not colors:", difference_result)  # {'grapes', 'apples'}
```

## DICTIONARY

### Create a dictionary d = {"actor":"amir","animal":"cat","earth":2,"list":[23,32,12]}

```python
# Creating a dictionary with key-value pairs
# Dictionaries use curly braces {} with key:value format
d = {"actor": "amir", "animal": "cat", "earth": 2, "list": [23, 32, 12]}
print(d)
```

Print the value of d[0]

```python
# Dictionaries use KEYS to access values, not numeric indices
# There is no key '0' in this dictionary, so it will raise KeyError
print(d[0])  # KeyError: 0
```

<b>Oops!! again an error. again a fun fact. Dictionary return the value for key if key is in the dictionary, else throws KeyError
and we don't have key 0 here :( </b>

Store the value of d[‚Äòactor‚Äô] to a new variable actor.

```python
# Accessing dictionary values using keys
# d['actor'] returns the value associated with key 'actor'
actor = d['actor']
print(actor)  # Output: amir
```

Print the type of actor

```python
# Checking the type of the value stored in 'actor'
print(type(actor))  # Output: <class 'str'> (it's a string)
```

Store the value of d[‚Äòlist‚Äô] in new variable List. 

```python
# Dictionary values can be any data type, including lists
# Storing the list value in a new variable
List = d['list']
print(List)  # Output: [23, 32, 12]
```

Print the type of List. 

```python
# The value stored was a list
print(type(List))  # Output: <class 'list'>
```

Create d1 = { ‚Äòsinger‚Äô : ‚ÄòKr$na‚Äô , ‚Äòalbum‚Äô: ‚ÄòStill here‚Äô, ‚Äògenre‚Äô : ‚Äòhip-hop‚Äô}

```python
# Creating another dictionary with music information
d1 = {'singer': 'Kr$na', 'album': 'Still here', 'genre': 'hip-hop'}
print(d1)
```

Merge d1 into d.

```python
# update() method merges another dictionary into the current one
# All key-value pairs from d1 are added to d
d.update(d1)
print("Merged dictionary:", d)
```

print d

```python
# Printing the merged dictionary
print(d)
```

Print all the keys in d

```python
# keys() method returns all keys in the dictionary
print(d.keys())
# Output: dict_keys(['actor', 'animal', 'earth', 'list', 'singer', 'album', 'genre'])
```

Print all the values in d

```python
# values() method returns all values in the dictionary
print(d.values())
# Output: dict_values(['amir', 'cat', 2, [23, 32, 12], 'Kr$na', 'Still here', 'hip-hop'])
```

Iterate over d, and print each key, value pair as this - (actor ----> amir)

```python
# items() method returns key-value pairs as tuples
# We can iterate using for loop with two variables
for key, value in d.items():
    print(f"{key} ----> {value}")
```

count the number of occurences of charachters in string named "sent" using dictionary and print the same.


```python
# Counting character occurrences in a string using dictionary
# Using the 'sent' variable from earlier
sent = 'Coronavirus Caused Lockdowns Around The World.'

# Create an empty dictionary to store counts
char_count = {}

# Iterate through each character in the string
for char in sent:
    # If character already exists, increment count
    if char in char_count:
        char_count[char] += 1
    # If character is new, add it with count 1
    else:
        char_count[char] = 1

# Print the character count dictionary
print("Character counts:")
print(char_count)
```
