**Purpose:**  
Load the wine dataset, compute central tendency measures (mean, median, mode), create a representative summary, and plot feature distributions. Code includes comments to explain the flow.


```python
# Imports and configuration
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Path to dataset (adjust if needed)
DATA_PATH = 'Wine Quality Dataset.csv'

# Load dataset into a DataFrame
df = pd.read_csv(DATA_PATH)

# Show the first 5 rows to understand structure
df.head()
```

```python
# Calculate central tendency measures for numeric columns
# mean, median, and the most frequent value (mode)
mean_values = df.mean(numeric_only=True)
median_values = df.median(numeric_only=True)
mode_values = df.mode(numeric_only=True).iloc[0]

# Combine into a single DataFrame for clear comparison
summary = pd.DataFrame({
    'Mean': mean_values,
    'Median': median_values,
    'Mode': mode_values
})

# Display the summary table
summary
```

```python
# Create a representative value for each feature
# Rule used (simple and explainable): 
# if absolute skewness > 0.5 -> use median (data likely skewed)
# else -> use mean (data roughly symmetric)
skewness = df.skew(numeric_only=True)

representative = []
for col in summary.index:
    s = skewness.get(col, 0)
    if abs(s) > 0.5:
        representative.append(median_values[col])
    else:
        representative.append(mean_values[col])

representative_series = pd.Series(representative, index=summary.index, name='Representative')

# Show the representative series
representative_series
```

```python
# Save the summary and representative series for later use
summary.to_csv('wine_central_tendency_summary.csv')
representative_series.to_csv('wine_representative_series.csv', header=True)

print('Saved:')
print('wine_central_tendency_summary.csv')
print('wine_representative_series.csv')
```

```python
# Plot distributions for numeric features
# Each plot is a simple histogram with labels and title.
numeric_cols = df.select_dtypes(include='number').columns.tolist()

for col in numeric_cols:
    plt.figure(figsize=(10,5))
    plt.hist(df[col].dropna(), bins=30)
    plt.title(f'Distribution of {col}', fontsize=16)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(axis='y', linestyle='--', linewidth=0.5)
    plt.show()
```

This project analyzes the **Wine Quality dataset** to understand how its chemical features behave using **central tendency measures**. The dataset was loaded into a DataFrame, and each numeric feature was examined by calculating its **mean**, **median**, and **mode**. A representative value for each feature was then selected based on its distribution.

To better understand the data patterns, **histograms** were plotted for all variables, giving a clear view of their spread and skewness. Finally, a summary table and a representative-value series were created for quick reference.
