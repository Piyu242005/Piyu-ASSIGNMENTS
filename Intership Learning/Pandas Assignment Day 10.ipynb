## Applying functions to columns and creating new columns

We need to create another column in d3, which contains  a boolean value for each city to indicate whether it's a union territory or not. 
- HINT: Chandigarh, Pondicherry and Delhi are only 3 union territories here. 


```python
# Function to check if a city is a Union Territory (UT)
# Returns True if the city is Chandigarh, Delhi, or Pondicherry
def is_ut(x):
    # List of Union Territories in our dataset
    union_territories = ['Chandigarh', 'Delhi', 'Pondicherry']
    # Return True if city is in the list, else False
    return x in union_territories

# Apply the function to the 'city' column to create a new column 'is_ut'
d3['is_ut'] = d3['city'].apply(is_ut)
```

```python
# Print d3 to see the new 'is_ut' column
print("Dataframe d3 with is_ut column:")
print(d3)
```

The temperatures mentioned in ‘Temperature’ column are mentioned in Celsius, we need another column which contains the same in Fahrenheit. 

HINT - 
- Define a function c_to_f which takes input temp in celsius and returns a value with temperature in Fahrenheit.
- To check: c_to_f(10) should return 50. 


```python
# Function to convert Celsius to Fahrenheit
# Formula: F = (C * 9/5) + 32
def c_to_f(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit
```

```python
# Test the c_to_f function with 10°C
# Expected output: 50°F
print("c_to_f(10) =", c_to_f(10))
```

```python
# Apply the c_to_f function to create a new column 'temp_fahrenheit'
# apply() method applies a function along an axis of the DataFrame
d3['temp_fahrenheit'] = d3['Temperature'].apply(c_to_f)
print("Dataframe d3 with temp_fahrenheit column:")
print(d3)
```

## Groupby

```python
# Create a sample dataframe for demonstrating groupby operations
# Contains sales data with product categories, regions, and values
data = pd.DataFrame({
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Food', 'Clothing', 'Food', 'Electronics', 'Clothing'],
    'Region': ['North', 'South', 'North', 'East', 'North', 'South', 'East', 'East'],
    'Sales': [1500, 800, 2000, 300, 950, 450, 1800, 700],
    'Quantity': [10, 20, 15, 50, 25, 60, 12, 30]
})
print("Sample dataframe 'data':")
print(data)
```

```python
# Groupby single column 'Category' with aggregate sum()
# This calculates the total sales and quantity for each category
print("Groupby Category with sum():")
print(data.groupby('Category').sum())
```

```python
# Groupby single column 'Category' with aggregate count()
# This counts the number of rows for each category
print("Groupby Category with count():")
print(data.groupby('Category').count())
```

```python
# Groupby single column 'Category' with min() and max()
# Using agg() to apply multiple aggregate functions
print("Groupby Category with min() and max():")
print(data.groupby('Category').agg(['min', 'max']))
```

```python
# Groupby two columns 'Category' and 'Region' with mean()
# This calculates the average sales and quantity for each category-region combination
print("Groupby Category and Region with mean():")
print(data.groupby(['Category', 'Region']).mean())
```

```python
# Groupby two columns 'Category' and 'Region' with min() and max()
print("Groupby Category and Region with min() and max():")
print(data.groupby(['Category', 'Region']).agg(['min', 'max']))
```

```python
# Additional: Display summary statistics
print("Summary statistics:")
print(data.describe())
```
