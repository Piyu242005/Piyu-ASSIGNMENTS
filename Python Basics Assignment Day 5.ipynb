# **`Logical Operators`**

```python
# compare if 108 is equall to 108, 21 is equall to 21 using logical and 
# equall to => '=='
# logical and => and

# in and both condition must be True to get a True

# Logical AND: Both conditions must be True for result to be True
result = (108 == 108) and (21 == 21)
print(result)  # Output: True (both conditions are True)
```

```python
# how above condition can give False as output show all those conditions

# For AND to return False, at least one condition must be False
print((108 == 108) and (21 == 22))  # False - second condition is False
print((108 == 107) and (21 == 21))  # False - first condition is False
print((108 == 107) and (21 == 22))  # False - both conditions are False
```

```python
# compare if 108 is equall to 108, 21 is equall to 11 using logical or
# equall to => '=='
# logical or => or

# in or Only one condition need to be True to get a True

# Logical OR: At least one condition must be True for result to be True
result = (108 == 108) or (21 == 11)
print(result)  # Output: True (first condition is True, that's enough for OR)
```

```python
# assign money variable value of 10000
##### say you have this much ammount in your account

money = 10000  # Assigning value to money

# start of if condition
# if money is greater then 1000 which is data science course free
# if money > 1000 is false i.e. you have less money then 1000 in your account then else will work for now only if is working

# if-else statement: executes different blocks based on condition
if money > 1000:
    # This block runs if condition is True
    print("You can afford the Data Science course!")
    print("Your balance:", money)
else:
    # This block runs if condition is False
    print("Sorry, you don't have enough money for the course")
```

```python
# take a test_score variable with 80 in it.

test_score = 80  # Student's test score

# if test_score greater then 80 then print A grade
# elif test_score greater then 60 and less then 80 print B grade
# else print Nothing for you

# if-elif-else ladder for multiple conditions
if test_score > 80:
    # First condition: score above 80
    print("A grade")
elif test_score > 60 and test_score <= 80:
    # Second condition: score between 61-80
    print("B grade")
else:
    # Default case: score 60 or below
    print("Nothing for you")
```

# **`Break Statement`**


```python
# example that uses break statement in a for loop

# take range(10) and print 'The number is' + value
# break when num equals 5

# break statement exits the loop completely
for num in range(10):
    if num == 5:
        # When num reaches 5, exit the loop
        break
    # This prints for num = 0, 1, 2, 3, 4
    print('The number is', num)
```

# **`Continue Statement`**


```python
# Using same `for loop program` as in Break Statement section above
# Use a continue statement rather than a break statement

# take range(10) and print 'The number is' + value
# continue when num equals 5

# continue statement skips current iteration and moves to next
for num in range(10):
    if num == 5:
        # Skip printing when num is 5
        continue
    # This prints for all numbers except 5
    print('The number is', num)
```

# ***`Positional Arguments`*** 
Most arguments are identified by their position in function call
* Say `print(x,y)` will give different results from `print(y,x)`

What ever sequence is given while defining a function values must be taken in that sequence only
* Otherwise use argument name **`(keyword arguments)`** to take values
* We first define `positional argument` and then `keyword arguments`


```python
## Create substraction_function(small_number,large_number) and return difference between large_number and small_number

# Function with positional parameters
def substraction_function(small_number, large_number):
    """
    Subtracts small_number from large_number
    Returns the difference
    """
    difference = large_number - small_number
    return difference
```

```python
# pass arguments in right order

# Positional arguments: order matters!
# First argument goes to small_number, second to large_number
result = substraction_function(5, 10)  # 10 - 5 = 5
print(result)  # Output: 5
```

```python
# always pass arguments using there name(keyword arguments) then order does not matter

# Keyword arguments: specify parameter names, order doesn't matter
result = substraction_function(large_number=20, small_number=8)
print(result)  # Output: 12 (20 - 8)

# Another example with different order
result2 = substraction_function(small_number=3, large_number=15)
print(result2)  # Output: 12 (15 - 3)
```
